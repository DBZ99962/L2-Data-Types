### 1.

Write a Python program to perform following tasks on a list

<ol>
    <li>Create an empty list with name l1</li>
    <li>Add atleast 5 number to the l1</li>
    <li>Add 100 at 2<sup>nd</sup> position</li>
    <li>Add 1000 at front of the l1</li>
    <li>Delete element at 3<sup>rd</sup> position</li>
    <li>Display elements at even position</li>
    <li>Display elements in reverse order</li>
</ol>

```python
# Question 1: List operations
# Step 1: Create an empty list named l1
l1 = []

# Step 2: Add atleast 5 numbers to the l1
l1.append(10)
l1.append(20)
l1.append(30)
l1.append(40)
l1.append(50)

# Step 3: Add 100 at 2nd position (index 1)
l1.insert(1, 100)

# Step 4: Add 1000 at front of the l1 (at index 0)
l1.insert(0, 1000)

# Step 5: Delete element at 3rd position (index 2, since list is now 0-indexed)
l1.pop(2)

# Display current list
print("Current list:", l1)

# Step 6: Display elements at even position
print("\nElements at even positions:")
for i in range(0, len(l1), 2):
    print(f"Position {i}: {l1[i]}")

# Step 7: Display elements in reverse order
print("\nElements in reverse order:")
for element in reversed(l1):
    print(element)
```

### 2. 

Grab ``StarAgile`` from the following Dictionaries

d_difficult = {'k1':[{'nest_key':['this is deep',['StarAgile']]}]}

d_irritating = {'k1':[1,2,{'k2':['this is not easy',{'VeryIrriating':[1,2,['StarAgile']]}]}]}

```python
# Question 2: Extract StarAgile from nested dictionaries
# Defining the two dictionaries
d_difficult = {'k1':[{'nest_key':['this is deep',['StarAgile']]}]}
d_irritating = {'k1':[1,2,{'k2':['this is not easy',{'VeryIrriating':[1,2,['StarAgile']]}]}]}

# Extracting StarAgile from d_difficult by navigating through the nested structure
# d_difficult['k1'][0] accesses the first dictionary in the list under 'k1'
# ['nest_key'][1][0] accesses the 'StarAgile' string inside the nested list
star_agile_difficult = d_difficult['k1'][0]['nest_key'][1][0]
print(f"StarAgile from d_difficult: {star_agile_difficult}")

# Extracting StarAgile from d_irritating by navigating through the nested structure
# d_irritating['k1'][2] accesses the dictionary at index 2 in the list under 'k1'
# ['k2'][1]['VeryIrriating'][2][0] navigates to 'StarAgile' string
star_agile_irritating = d_irritating['k1'][2]['k2'][1]['VeryIrriating'][2][0]
print(f"StarAgile from d_irritating: {star_agile_irritating}")

# Verifying both values are the same
print(f"\nBoth values are same: {star_agile_difficult == star_agile_irritating}")
print(f"Final Answer: {star_agile_difficult}")
```

### 3.

Write a program to accept a string from the user and display characters that are present at an even index number.

For example, str = "judgemental" so you should display ‘j’, ‘d’, ‘e’, ‘e’, ‘t’, ‘l’.

```python
# Question 3: Extract characters at even index numbers
# Accept a string from user
user_string = input("Enter a string: ")

# Extract and display characters at even index positions (0, 2, 4, 6, ...)
print("\nCharacters at even index positions:")

# Method 1: Using string slicing (most efficient)
# string[start:end:step] where step=2 gets every 2nd character starting from index 0
even_index_chars = user_string[::2]
print(f"Using slicing: {even_index_chars}")

# Method 2: Using loop to demonstrate the logic
print("\nCharacters displayed one by one:")
for i in range(0, len(user_string), 2):
    # For each even index position, print the character
    print(f"Index {i}: '{user_string[i]}'")

# Method 3: Display all even-indexed characters in a comma-separated format
print(f"\nAll even-indexed characters: {', '.join([user_string[i] for i in range(0, len(user_string), 2)])}")

# Example demonstration with the given example
print("\n--- Example Demonstration ---")
example_str = "judgemental"
print(f"Example String: {example_str}")
print(f"Characters at even indices: {', '.join([example_str[i] for i in range(0, len(example_str), 2)])}")
print(f"Or displayed as: {example_str[::2]}")
# This should output: j, d, e, e, t, l
```

### 4. 

Convert the Float Value of '3.8' into Integer, Complex, Binary and Hexidecimal format

```python
# Question 4: Convert float value of 3.8 into different formats
# Define the float value
float_value = 3.8

print(f"Original Float Value: {float_value}")
print("\n" + "="*50)

# Convert to Integer (truncates decimal part)
integer_value = int(float_value)
print(f"\n1. Integer Format: {integer_value}")
print(f"   Explanation: int() truncates the decimal part, 3.8 becomes {integer_value}")

# Convert to Complex number (adds imaginary part)
complex_value = complex(float_value, 0)
print(f"\n2. Complex Format: {complex_value}")
print(f"   Explanation: complex(real, imag) creates a complex number")

# Convert to Binary format
binary_value = bin(int(float_value))
print(f"\n3. Binary Format: {binary_value}")
print(f"   Explanation: bin() converts the integer part to binary representation")
print(f"   (Note: 3 in decimal = 11 in binary)")

# Convert to Hexadecimal format
hex_value = hex(int(float_value))
print(f"\n4. Hexadecimal Format: {hex_value}")
print(f"   Explanation: hex() converts the integer part to hexadecimal representation")
print(f"   (Note: 3 in decimal = 3 in hexadecimal)")

# Summary
print("\n" + "="*50)
print("\nSUMMARY of conversions for 3.8:")
print(f"Integer: {integer_value}")
print(f"Complex: {complex_value}")
print(f"Binary: {binary_value}")
print(f"Hexadecimal: {hex_value}")

# Additional examples to show the conversion process
print("\n" + "="*50)
print("\nDetailed Conversion Process:")
print(f"3.8 → int() → {integer_value}")
print(f"3.8 → complex() → {complex_value}")
print(f"int(3.8)=3 → bin() → {binary_value}")
print(f"int(3.8)=3 → hex() → {hex_value}")
```

### 5. 

Write a program to accept five names from the user and store it in a Tuple 'T1'

```python
# Question 5: Accept five names from user and store in a Tuple T1
# Initialize an empty list to temporarily store names
names_list = []

# Prompt user to enter five names
print("Welcome! Please enter five names one by one.")
print("=" * 50)

# Loop to accept 5 names from user
for i in range(1, 6):
    # Accept name from user with a prompt
    name = input(f"Enter name {i}: ")
    # Append the name to the list
    names_list.append(name)

# Convert the list to a tuple (tuples are immutable, fixed collections)
# Tuples are created by passing the list to tuple() function
T1 = tuple(names_list)

# Display the tuple
print("\n" + "=" * 50)
print(f"\nTuple T1 containing the five names: {T1}")
print(f"\nData type of T1: {type(T1)}")

# Display each name individually
print("\n" + "=" * 50)
print("\nNames in T1 (accessing individually by index):")
for i, name in enumerate(T1):
    # Display each name with its index position (0-indexed)
    print(f"T1[{i}] = {name}")

# Show tuple properties
print("\n" + "=" * 50)
print("\nTuple Properties:")
print(f"Length of T1: {len(T1)}")
print(f"First name: {T1[0]}")
print(f"Last name: {T1[-1]}")
print(f"\nTuple is immutable - meaning once created, it cannot be modified")
```
